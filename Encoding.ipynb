```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Encoding Data.csv')
```

```python
df.head(10)
```

#label encoding

```python
#applying Label encoding in ord_2 column(label)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['ord_2']=le.fit_transform(df['ord_2'])
```

```python
df
```

#one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder()
#transforming the columns after fitting
enc=enc.fit_transform(df[['nom_0']]).toarray()
#conveting array to a Dataframe
encoded_colm=pd.DataFrame(enc)
#concating DataFrames
df=pd.concat([df,encoded_colm],axis=1)
#removing the encoded column
df=df.drop(['nom_0'],axis=1)
df.head(10)
```
